```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
## Missing values
```

```python
feature_with_na =[features for features in dataset.columns if dataset[features].isnull().sum()>1]
```

```python
print("Count with features as Null - {}/{}".format(len(feature_with_na),len(dataset.columns)))
```

```python
for feature in feature_with_na :
    print("Feature with Na percent {} - {} % ".format(feature,np.round(np.mean(dataset[feature].isnull()*100),2)))
```

```python

for feature in feature_with_na:
    data = dataset.copy()
    
    data[feature] =np.where(data[feature].isnull(),1,0)
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
    
```

```python
## Numerical values - Continuous and Discrete numerical values
```

```python
features_numerical= [features for features in dataset.columns if dataset[features].dtype!='O']
```

```python
print("Count of Numerical features - {}/{} ".format(len(features_numerical),len(dataset.columns)))
```

```python
## Temporal Variables
```

```python
year_feature = [feature for feature in features_numerical if 'Yr' in feature or 'Year' in feature]
```

```python
year_feature
```

```python
for feature in year_feature:
    data = dataset.copy()
    if feature != 'YrSold' :
        data[feature] =data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('YrSold')
        plt.show()
```

```python
## Numerical and Categorical Features
```

```python
## Continuous Numerical
```

```python
discrete_nummerical = [feature for feature in features_numerical if len(dataset[feature].unique())< 25 and feature not in year_feature]
    
```

```python
len(discrete_nummerical)
```

```python
discrete_nummerical
```

```python
for feature in discrete_nummerical:
    data = dataset.copy()
    
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.show()
```

```python
continuous_feature =[feature for feature in features_numerical if feature not in discrete_nummerical+year_feature+['Id']]
```

```python
len(continuous_feature)
```

```python
continuous_feature
```

```python
dataset['LotFrontage'].unique()
```

```python
for feature in continuous_feature:
    
    data= dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.show()
```

```python
for feature in continuous_feature:
    
    data= dataset.copy()
    
    data[feature]= np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.show()
```

```python
## Outliers
```

```python
for feature in continuous_feature:
    
    data= dataset.copy()
    data[feature]= np.log(data[feature])
    data.boxplot(column = feature)
    plt.show()
```
