```python
## Handle the Numerical features with median value
```

```python
numerical_nan = [feature for feature in dataset.columns if dataset[feature].dtype!='O' and dataset[feature].isnull().sum()>0]
```

```python
dataset[numerical_nan].isnull().head()
```

```python
for feature in numerical_nan :   
    print("Null of numerical feature {} is - {} % ".format(feature , np.round(dataset[feature].isnull().mean()*100,2)))
```

```python
for feature in numerical_nan:
    
    median_value = dataset[feature].median()
    dataset[feature+'_nan']= np.where(dataset[feature].isnull(),1,0)
    
    dataset[feature].fillna(median_value,inplace=True)
```

```python
## Temporal Variables change to no of years
```

```python
temporal_feature =[feature for feature in dataset.columns if dataset[feature].dtype!='O' and ('Yr' in feature or 'Year' in feature)]
```

```python
temporal_feature
```

```python
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    dataset[feature] =dataset['YrSold']-dataset[feature]
```

```python
dataset.head()
```

```python
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
```

```python
for feature in num_features:
    dataset[feature] = np.log(dataset[num_features])
```

```python
dataset.head()
```

```python
## Handling Rare categorical features < 1% whose values
```

```python
categorical_feature = [feature for feature in dataset.columns if dataset[feature].dtype=='O']
```

```python
for feature in categorical_feature:
    count= dataset.groupby(feature)['SalePrice'].count()/len(dataset)
    temp_df= count[count <0.01].index
    dataset[feature] = np.where(dataset[feature].isin(temp_df),'Rare_val',dataset[feature])
```

```python
dataset.head()
```

```python
## enumerate - replacing all the categorical features into numerical features
```

```python
for feature in categorical_feature:
    
    labels_ordered = dataset.groupby(feature)['SalePrice'].mean().sort_values().index
    labels = {k:i for i,k in enumerate(labels_ordered,0)}
    dataset[feature]= dataset[feature].map(labels)
```
